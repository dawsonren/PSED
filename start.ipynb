```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("PolySi_GBangle.xlsx - Cleaned.csv")
```

```python
# plot CSL (categorical) vs Average Thermal Resistance (numerical) for each CSL category
plt.figure(figsize=(10, 6))
csl_categories = sorted(data['CSL'].unique())
plt.boxplot([data[data['CSL'] == csl]['Thermal Resistance'] for csl in csl_categories], tick_labels=csl_categories)
plt.xlabel('CSL')
plt.ylabel('Average Thermal Resistance')
plt.title('CSL vs Average Thermal Resistance')
plt.grid(True)
plt.show()
```
